**Q1-Explain how high-level languages like Python are converted into machine-understandable code, including the difference between compiled and interpreted languages.**

High-level languages like Python are designed to be easy for humans to read and write. However, computers only understand machine code (binary instructions made of 0s and 1s). So, the code we write must be translated into machine-understandable instructions.
The source code is converted into machine code using either: Compiler or Interpreter
Compiled Languages :- A compiler translates the entire program at once into machine code before execution.
Interpreted Languages :-An interpreter translates and executes the code line by line during runtime.
Python is mainly an interpreted language, but internally it also uses a small compilation step.

Key Difference Between Compiled and Interpreted Languages
Feature	Compiled Language	Interpreted Language
Translation	Entire program at once	Line by line
Speed	Faster	Slower
Error Detection	After compilation	During execution
Output	Executable file	No separate executable
Examples	C, C++	Python, JavaScript
